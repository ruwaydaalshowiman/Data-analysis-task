# Table 1

####  Getting to Know the Data by Displaying Data Types and shapes and printing the head of data frame , dropping the null Columns

```python
Table1_df = pd.read_excel (r'Table1.xlsx')
Table1_df = Table1_df.loc[:, ~Table1_df.columns.str.contains('^Unnamed')]

print("Before dropping null columns...\n")
# print(Table1_df.info())
print(Table1_df.shape)
Table1_df=Table1_df.dropna(axis=1, how='all')
print("After droppingnull columns...\n")
# print(Table1_df.info())
print(Table1_df.shape)

Table1_df.head()
```

#### dropping the null rows and displaying the number of rows and columns before and after , remove newlines in strings with space 

```python
print("Before dropping null rows...\n")
print(Table1_df.shape)
Table1_df.dropna(axis = 0, how = 'all', inplace = True)
print("After dropping null rows...\n")
print(Table1_df.shape)
Table1_df = Table1_df.replace(r'\\n',  ' ', regex=True) 
Table1_df.tail(50)
```

#### remove row that has 10 or more missing value

```python
print("Before dropping null rows...\n")
print(Table1_df.shape)

Table1_df = Table1_df.dropna( thresh=10 )
Table1_df = Table1_df.fillna('None')

print("After dropping null rows...\n")
print(Table1_df.shape)
```

#### displying the count of each unique value in all columns

```python
print('\n\n########## RECEIVER NAME ###############\n',Table1_df['RECEIVER_CORPORATE_BRANCH_NAME_EN'].value_counts())
print('\n\n########## RECEIVER ID ###############\n',Table1_df['RECEIVER_CORPORATE_BRANCH_ID'].value_counts())

print('\n\n########## SEND NAME ###############\n',Table1_df['SEND_CORPORATE_BRANCH_NAME_AR'].value_counts())
print('\n\n########## SEND ID ###############\n' , Table1_df['SEND_CORPORATE_BRANCH_ID'].value_counts())

print('\n\n############# AIR WAYBILL TYPE ############\n', Table1_df['AIR_WAYBILL_TYPE'].value_counts())
print('\n\n############# AIR WAYBILL STATUS ############\n',Table1_df['AIR_WAYBILL_STATUS'].value_counts())

print(Table1_df['DEPARTURE'].value_counts(),'\n\n#########################\n')
```

#### fill missing ID columns with zeros

```python
Table1_df[['RECEIVER_CORPORATE_BRANCH_ID']] = Table1_df[['RECEIVER_CORPORATE_BRANCH_ID']].fillna(0)
Table1_df[['SEND_CORPORATE_BRANCH_ID']] = Table1_df[['RECEIVER_CORPORATE_BRANCH_ID']].fillna(0)
```
