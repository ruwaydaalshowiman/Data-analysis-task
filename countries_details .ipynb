```python
import pandas as pd
```

# Reading Countries files 

#### Getting to Know the Data by Displaying Data Types and shapes and printing the head of data frame

```python
Countries_df = pd.read_excel (r'Countries.xlsx')
Countries_df.drop(['COUNTRY_ID','SUB_CONTINENT_ID'],axis='columns', inplace=True)
print(Countries_df.shape)
print(Countries_df.info())
Countries_df.head()
```

# Reading Cities files

#### Getting to Know the Data by Displaying Data Types and shapes and printing the head of data frame

```python
Cities_df = pd.read_excel (r'Cities.xlsx')
Cities_df.drop(['COUNTRY_ID','CITY_ID'],axis='columns', inplace=True)
print(Cities_df.shape)
print(Cities_df.info())
Cities_df.head()
```

# Merging both Countries data frame and Cities data frame

#### the reason behind this step is to get a Hierarchy of each country by getting the CONTINENT with its countries along with its cities

```python
Countries_AND_Cities = pd.merge(Countries_df,Cities_df[['CITY_NAME_AR','CITY_NAME_EN','REGION_NAME_AR','REGION_NAME_EN','COUNTRY_CODE']],how = 'outer',left_on=['COUNTRY_CODE'],right_on=['COUNTRY_CODE'])
print(Countries_AND_Cities.shape)
Countries_AND_Cities.head()
```

### dealing with null values

```python
Countries_AND_Cities[['CITY_NAME_EN','CITY_NAME_AR','REGION_NAME_AR','REGION_NAME_EN']] = Countries_AND_Cities[['CITY_NAME_EN','CITY_NAME_AR','REGION_NAME_AR','REGION_NAME_EN']].fillna("None")

Countries_AND_Cities
```

```python
Countries_AND_Cities[Countries_AND_Cities.isna().any(axis=1)]
```

```python
Countries_AND_Cities[['SUB_CONTINENT_CODE']] = Countries_AND_Cities[['SUB_CONTINENT_CODE']].fillna("None")
Countries_AND_Cities[Countries_AND_Cities.isna().any(axis=1)]
```

# Reading Airports files

#### Getting to Know the Data by Displaying Data Types and shapes and printing the head of data frame

```python
Airports_df = pd.read_excel (r'Airports.xlsx')
Airports_df.drop('PK_TRACKING_AIRPORT_ID',axis='columns', inplace=True)
print(Airports_df.shape)
print(Airports_df.info())
Airports_df.head()
```

### getting the countries information by joining the Airports data frame with Countries_AND_Cities data frame

```python
Airports_df_New = pd.merge(Airports_df , Countries_df[['COUNTRY_NAME_AR','COUNTRY_NAME_EN','CONTINENT_NAME_AR','CONTINENT_NAME_EN','COUNTRY_CODE']] ,how = 'left',left_on=['COUNTRY_CODE'],right_on=['COUNTRY_CODE'])
print(Airports_df_New.shape)

Airports_df_New.head()

```
